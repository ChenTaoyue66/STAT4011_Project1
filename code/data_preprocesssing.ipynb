```python
df=pd.read_csv('/Users/alysonmak/Desktop/dropcols_data.csv',header=None,names=['city','region','area_around','selling_point', 'total _price', 'unit-price', 'room', 'floor', 'total_floor','area', 'configuration', 'internal_area','building_type', 'orientation', 'buildin_struture', 'decoration', 'ladder_ratio', 'centralized_heating', 'release_time' ,'trade_type', 'last_trade_time', 'residence_type', 'residence_age', 'property_right', 'mortgage_info' ,'house_certification', 'listing_check_code', 'community_name', 'sub_district', 'visittime']).astype("str")
#打開剛剛轉存的csv文件，加標題
df = df.iloc[1: , :]
```

```python
name=['city','region','area_around','selling_point', 'total _price', 'unit-price', 'room', 'floor', 'total_floor','area', 'configuration', 'internal_area','building_type', 'orientation', 'buildin_struture', 'decoration', 'ladder_ratio', 'centralized_heating', 'release_time' ,'trade_type', 'last_trade_time', 'residence_type', 'residence_age', 'property_right', 'mortgage_info' ,'house_certification', 'listing_check_code', 'community_name', 'sub_district', 'visittime']
for name in name:#del all symbol
    df[name]=df[name].astype(str).str.replace(r'[^\w\s]+', '')
    df[name]=df[name].astype(str).str.replace('n', '')
    df[name]=df[name].astype(str).str.replace('xa0', '')
```

```python
df.to_excel('dropsymbol_data.xlsx')
```
